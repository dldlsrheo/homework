```python
import matplotlib.animation as manimation
import h5py
import pandas as pd
import os
import tables
import numpy as np
from matplotlib import pyplot as plt
```

```python
h5file = tables.open_file(f'twiss.h5','r')
```

```python
s = h5file.root.s.read()
```

```python
alpha_x = h5file.root.alpha_x.read()
beta_x = h5file.root.beta_x.read()
D_x = h5file.root.D_x.read()
```

```python
alpha_x
```

```python
fig = plt.figure(figsize = (20,20))
```

```python
plt.figure(figsize = (10,10))
wave1 = plt.plot(s, alpha_x, 'b',label= r'$alpha_x$')
wave2 = plt.plot(s, beta_x, 'r',label= r'$beta_x$')
wave3 = plt.plot(s, D_x, 'g',label= r'$D_x$')
plt.xlabel('s[m]', fontsize=16)
plt.ylabel('x', fontsize=16)
plt.title('alpha_x,beta_x, Dispersion_x')
plt.legend(loc=2, fontsize=16)
plt.grid(1)
```
